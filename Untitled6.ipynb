```python
import pandas as pd #data manipulation
import numpy as np #mathematical operations
import matplotlib.pyplot as plt #graph plotting
```

```python
#create the data
student_details={
    "Names":["Ann","Peter","Douglas","Betty","Blick"],
    "Scores":[90,40,45,60,72]
}
print(student_details)
```

```python
#create a dataframe of the data
df=pd.DataFrame(student_details)
df
```

```python
#create a barplot to display the data
plt.bar(df['Names'],df['Scores'],color='g')
plt.title("Students Scores")
plt.xlabel("Students Names")
plt.ylabel("Student Scores")
plt.show()
```

#Findings


*   Ann is the highest performing student because she has the longest bar
*  Peter is the lowest performing student because he has the shortest bar



```python
#line chart
plt.plot(df['Names'],df['Scores'],color='b')
plt.title("Students Scores")
plt.xlabel("Students Names")
plt.ylabel("Student Scores")
plt.show()
```

#Findings


*   The students scores are dropping and increasing from Peter
